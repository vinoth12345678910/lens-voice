```python
from dotenv import load_dotenv
import os
import requests
import base64

load_dotenv()
api_key = os.getenv("SARVAM_API_KEY").strip()

url = "https://api.sarvam.ai/text-to-speech"
headers = {
    "api-subscription-key": api_key,
    "Content-Type": "application/json"
}
payload = {
    "inputs": ["Hello, this is a test."],
    "target_language_code": "en-IN",
    "speaker": "priya",
    "model": "bulbul:v3"
}

response = requests.post(url, headers=headers, json=payload)
print("Status:", response.status_code)

if response.status_code == 200:
    data = response.json()
    print("Response keys:", data.keys())
    audio_base64 = data["audios"][0]
    audio_bytes = base64.b64decode(audio_base64)
    with open("test_output.wav", "wb") as f:
        f.write(audio_bytes)
    print("Saved test_output.wav")
else:
    print(response.text)
```

```python
def speak(text_english, target_language="en-IN", speaker="priya", filename="output.wav"):
    """
    Takes an English sentence, translates if needed, and produces speech audio.
    target_language: "en-IN", "ta-IN", or "hi-IN"
    """
    final_text = text_english

    if target_language != "en-IN":
        translate_response = requests.post(
            "https://api.sarvam.ai/translate",
            headers=headers,
            json={
                "input": text_english,
                "source_language_code": "en-IN",
                "target_language_code": target_language
            }
        )
        if translate_response.status_code == 200:
            final_text = translate_response.json()["translated_text"]
        else:
            print("Translation failed, falling back to English:", translate_response.text)
            target_language = "en-IN"
            final_text = text_english

    tts_response = requests.post(
        "https://api.sarvam.ai/text-to-speech",
        headers=headers,
        json={
            "inputs": [final_text],
            "target_language_code": target_language,
            "speaker": speaker,
            "model": "bulbul:v3"
        }
    )

    if tts_response.status_code == 200:
        audio_bytes = base64.b64decode(tts_response.json()["audios"][0])
        with open(filename, "wb") as f:
            f.write(audio_bytes)
        return filename
    else:
        print("TTS failed:", tts_response.text)
        return None
```

```python
def generate_description(obj):
    """
    obj: a dict from either the urgency classifier (HAZARD) or change detector (INFO)
         expected keys: "class", "motion" (optional), "position" (optional), "distance" (optional), "urgency"
    Returns: a natural English sentence
    """
    cls = obj["class"]
    urgency = obj.get("urgency", "INFO")
    motion = obj.get("motion", "static")
    position = obj.get("position", "ahead")
    distance = obj.get("distance", "medium")

    # Friendly names for classes (raw IDD labels aren't always natural to say)
    friendly_names = {
        "car": "a car",
        "bus": "a bus",
        "truck": "a truck",
        "motorcycle": "a motorcycle",
        "autorickshaw": "an auto-rickshaw",
        "bicycle": "a bicycle",
        "person": "a person",
        "rider": "a rider",
        "animal": "an animal",
        "traffic light": "a traffic light",
        "traffic sign": "a traffic sign",
        "trailer": "a trailer",
        "caravan": "a caravan",
        "train": "a train",
        "vehicle fallback": "a vehicle",
    }
    subject = friendly_names.get(cls, cls)

    if urgency == "HAZARD":
        if motion == "approaching":
            return f"Careful, {subject} is approaching from your {position}."
        else:
            return f"Careful, {subject} is close on your {position}."

    # INFO lane phrasing
    distance_phrase = {
        "near": "close by",
        "medium": "nearby",
        "far": "further away"
    }.get(distance, "nearby")

    return f"There is {subject} {distance_phrase} on your {position}."
```

```python
test_objects = [
    {"class": "car", "urgency": "HAZARD", "motion": "approaching", "position": "left"},
    {"class": "chair", "urgency": "INFO", "position": "right", "distance": "far"},
    {"class": "person", "urgency": "INFO", "position": "center", "distance": "near"},
]

for obj in test_objects:
    sentence = generate_description(obj)
    print(sentence)
```

```python
def describe_and_speak(obj, target_language="en-IN", speaker="priya"):
    sentence = generate_description(obj)
    print(f"Speaking: {sentence}")
    filename = speak(sentence, target_language=target_language, speaker=speaker)
    return filename
```
